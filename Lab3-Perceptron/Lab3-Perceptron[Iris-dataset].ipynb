### Implementing the Perceptron

```python
class Perceptron(object):
    """Perceptron classifier.

    Parameters
    ------------
    eta : float
        Learning rate (between 0.0 and 1.0)
    n_iter : int
        Passes over the training dataset.

    Attributes
    -----------
    w_ : 1d-array
        Weights after fitting.
    errors_ : list
        Number of misclassifications in every epoch.

    """
    def __init__(self, eta=0.01, n_iter=10):
        self.eta = eta
        self.n_iter = n_iter

    def fit(self, X, y):
        """Fit training data.

        Parameters
        ----------
        X : {array-like}, shape = [n_samples, n_features]
            Training vectors, where n_samples is the number of samples and
            n_features is the number of features.
        y : array-like, shape = [n_samples]
            Target values.

        Returns
        -------
        self : object

        """
        self.w_ = np.zeros(1 + X.shape[1])
        self.errors_ = []

        for _ in range(self.n_iter):
            errors = 0
            for xi, target in zip(X, y):
                update = self.eta * (target - self.predict(xi))
                self.w_[1:] += update * xi
                self.w_[0] += update
                errors += int(update != 0.0)
            self.errors_.append(errors)
        return self

    def net_input(self, X):
        """Calculate net input"""
        return np.dot(X, self.w_[1:]) + self.w_[0]

    def activation(self, X):
        """Compute linear activation"""
        return self.net_input(X)

    def predict(self, X):
        """Return class label after unit step"""
        return np.where(self.activation(X) >= 0.0, 1, -1)
```

### Selecting the binary targets

```python
# Dictionary to hold the binary targets
y = {}

y['setosa-versicolor'] = df.iloc[0:100, 4].values
y['setosa-versicolor'] = np.where(y['setosa-versicolor'] == 'Iris-setosa', -1, 1)

y['versicolor-virginica'] = df.iloc[50:150, 4].values
y['versicolor-virginica'] = np.where(y['versicolor-virginica'] == 'Iris-versicolor', -1, 1)

y['virginica-setosa'] = np.concatenate((df.iloc[100:150, 4].values, df.iloc[0:50, 4].values))
y['virginica-setosa'] = np.where(y['virginica-setosa'] == 'Iris-virginica', -1, 1)
```

#### Visualising all the binary classes between all the features in 2 dimension space to find out which features are linearly separable 

```python
def visualize(X, class1, class2):
    """Function to plot the different classes in X for all the available features in 2-D space
    
    Parameter:
    ----------
    X: dataset containing the features of the flower present
    class1: name of the first class of flower in the dataset
    class2: name of the second class of flower in the dataset
    """
    
    features = ['sepal_length_cm', 'sepal_width_cm', 'petal_length_cm', 'petal_width_cm']
    combinations = [[0, 1], [1, 2], [2, 3], [3, 0]]
    
    for j in combinations:
        c1_x0 = X[0:50][:, j[0]]
        c1_x1 = X[0:50][:, j[1]]
        c2_x0 = X[50:100][:, j[0]]
        c2_x1 = X[50:100][:, j[1]]
        
        ax = plt.figure(figsize = (5, 4))
        plt.scatter(c1_x0, c1_x1, color = 'red', marker = 'o', label = class1)
        plt.scatter(c2_x0, c2_x1, color = 'blue', marker = 'x', label = class2)
        
        plt.title(class1 + ' - ' + class2)
        plt.xlabel(features[j[0]])
        plt.ylabel(features[j[1]])
        plt.legend(loc = 'upper left')
        plt.show()
```

```python
visualize(X['setosa-versicolor'], 'setosa', 'versicolor')
```

```python
visualize(X['versicolor-virginica'], 'versicolor', 'virginica')
```

```python
visualize(X['virginica-setosa'], 'virginica', 'setosa')
```

#### From the above visualisations, we find that Iris-setosa is linearly separable from all the other classes on the basis of all the features whereas 'Iris-versicolor' and 'Iris-virginica' are linearly separable only between 'petal_length_cm' and 'petal_width_cm' 

#### Selecting only the linearly separable features

## Training the perceptron 

```python
# Dictionary to hold all the perceptrons
ppn = {}

ppn['setosa-versicolor'] = Perceptron(eta = 0.01, n_iter = 100)
ppn['versicolor-virginica'] = Perceptron(eta = 0.01, n_iter = 1000)
ppn['virginica-setosa'] = Perceptron(eta = 0.01, n_iter = 100)
```

```python
ppn['setosa-versicolor'].fit(X['setosa-versicolor'], y['setosa-versicolor'])
ppn['versicolor-virginica'].fit(X['versicolor-virginica'], y['versicolor-virginica'])
ppn['virginica-setosa'].fit(X['virginica-setosa'], y['virginica-setosa'])
```

### Plotting the misclassifications during the training 

```python
for key, value in ppn.items():
    plt.plot(range(1, len(value.errors_) + 1), value.errors_, marker = '.')
    plt.title(key)
    plt.xlabel('Epochs')
    plt.ylabel('# misclassification in every epoch')
    plt.show()
```

## Visualising the perceptron

```python
from matplotlib.colors import ListedColormap
```

```python
def plot_decision_regions(X, y, classifier, resolution = 0.02):
    
    # setup marker generator and color map
    markers = ('s', 'x', 'o', '^', 'v')
    colors = ('red', 'blue', 'lightgreen', 'gray', 'cyan')
    cmap = ListedColormap(colors[:len(np.unique(y))])
    
    # plot the decision surface
    x1_min, x1_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    x2_min, x2_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx1, xx2 = np.meshgrid(np.arange(x1_min, x1_max, resolution),
                          np.arange(x2_min, x2_max, resolution))
    Z = classifier.predict(np.array([xx1.ravel(), xx2.ravel()]).T)
    Z = Z.reshape(xx1.shape)
    plt.contourf(xx1, xx2, Z, alpha = 0.4, cmap = cmap)
    plt.xlim(xx1.min(), xx1.max())
    plt.ylim(xx2.min(), xx2.max())
    
    # plotting class samples
    for idx, cl in enumerate(np.unique(y)):
        plt.scatter(x = X[y == cl, 0], y = X[y == cl, 1],
                   alpha = 0.8, c = cmap(idx), edgecolor = 'black',
                   marker = markers[idx], label = cl)
```

```python
for key, value in ppn.items():
    plot_decision_regions(X[key], y[key], classifier = value)
    plt.title(key)
    plt.xlabel('petal_width_cm')
    plt.ylabel('petal_length_cm')
    plt.legend(loc = 'upper left')
    plt.show()
```
